```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
df.head()
```

(i) Create an area plot for top 6 immigrant countries from 1990 to 2013

```python
k = df.sort_values(['Total'],ascending = False)
m = k.head(6)
farea = m.iloc[:,4:-1].transpose()
colm = m['Country']
farea.columns = colm
farea.plot.area()
```

(ii) Create an year-wise immigrant bar chart from India to Canada during the period of
	    1980 to 2013.


```python
xC = df.columns[4:-1]
yC = df.iloc[182,4:-1]
fig = plt.figure(figsize=(30,4))
plt.bar(xC,yC)
plt.title('Canada')
plt.show()
```

(iii) Create a boxplot for Indian, Phillipine and China immigrants.

```python
import seaborn as sns
col3 = ['India', 'Philippines', 'China']
new3 = pd.DataFrame()
for i,j in zip([0,1,3],col3) : new3[j] = df.iloc[i,4:-1]
fnew3 = new3.transpose()
fig,ax = plt.subplots()
plt.boxplot(fnew3)
```

(iv) Show the total no. of immigrants from India and France countries using Area Chart  and
	    Pie chart.


```python
new4 = pd.DataFrame()
for i,j in zip([0,11],['India','France']):
  new4[j] = df.iloc[i,4:-1]
  new4.plot
```

```python
fig1,ax1 = plt.sublots()
fnew4 = pd.DataFrame()
for i,j in zip([0,11],['India','France']):
  fnew4[j] = df.iloc[i,4:]
l4 = ['India','France']
pnew4 = fnew4.transpose()
ax1.pie(pnew4['Total'],labels = l4)
```

v .Creates a scatter Histogram for the immigrants from Fiji and Singapore in the year 2013

```python
import numpy as np
sns.set(color_codes = True)
new5df = pd.DataFrame({'Fiji' : [df.iloc[52,-2]],'Singapore': [df.iloc[66,-2]]})
sns.jointplot(x = "Fiji", y = "Singapore", data = new5df)
```
