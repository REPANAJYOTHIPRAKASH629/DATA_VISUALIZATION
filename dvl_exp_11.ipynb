```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = y = np.arange(-5, 6, 1)

# Create a meshgrid and calculate the corresponding z values
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Create a 3D heatmap using matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='plasma')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Heatmap')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x and y
x = y = np.arange(-5, 6, 1)

# Create a meshgrid and calculate the corresponding z values
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Create a 2D heatmap using matplotlib
plt.imshow(z, cmap='plasma', extent=(min(x[0]), max(x[0]), min(y[:,0]), max(y[:,0])))
plt.colorbar(label='z')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Heatmap')

plt.show()
```
