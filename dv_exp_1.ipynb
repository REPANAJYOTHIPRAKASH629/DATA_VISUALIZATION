# **BOX PLOT**

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(15)
dataSet = np.random.normal(100, 25, 200)
#print(dataSet)
figure = plt.figure(figsize =(5,5))
plt.boxplot(dataSet)
plt.show()
```

# **SCATTER PLOT**

```python
import pandas as pd
import matplotlib.pyplot as plot

data_value = [(3, 5),(25, 30),(7, 4),(14, 15)]
dataFrame = pd.DataFrame(data = data_value, columns = ['x_axis', 'y_axis']);
dataFrame.plot.scatter(x = 'x_axis', y = 'y_axis', title = "Scatter plot pandas variable for X and Y axis");
plot.show();
```

# **HISTOGRAM**

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]
plt.hist(data, bins=5, color='blue', edgecolor='black')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

# **PIE CHART**

```python
import matplotlib.pyplot as plt

data = [30, 25, 15, 10, 20]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
plt.title('Pie Chart DEMO')
plt.show()
```

# **FACET PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
g = sns.FacetGrid(data, col="species")
g.map(plt.scatter, "sepal_length", "sepal_width")
g.set_titles("{col_name}")
plt.show()
```

# **PAIR PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.pairplot(data, hue="species", markers=["o", "s", "D"])
plt.show()
```

# **AREA CHART**

```python
import matplotlib.pyplot as plt

x_data = [1, 2, 3, 4, 5]
y_data1 = [3, 4, 6, 8, 10]
y_data2 = [1, 2, 3, 4, 5]

plt.stackplot(x_data, y_data1, y_data2, labels=['Series 1', 'Series 2'], alpha=0.5)
plt.title('Stacked Area Chart Example')
plt.legend()
plt.show()
```

#**VIOLIN PLOT**


```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=data, palette="muted")
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Violin Plot Example')
plt.show()
```

#**BAR CHART**

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 15]
plt.bar(categories, values, color='cyan', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart Example')
plt.show()
```
