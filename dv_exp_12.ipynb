```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
(train_images, _), (_, _) = mnist.load_data()
num_samples = train_images.shape[0]
flat_images = train_images.reshape((num_samples, -1))
flat_images = flat_images / 255.0
n_components = 2  # Number of components to keep
pca = PCA(n_components=n_components)
reduced_features = pca.fit_transform(flat_images)
plt.figure(figsize=(8, 6))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c='black', marker='.')
plt.title('MNIST Dataset - Scatter Plot (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
