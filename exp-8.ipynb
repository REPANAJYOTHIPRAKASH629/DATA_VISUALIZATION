# WINE REVIEWS

# Problem Statement

# Import necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the dataset

```python
wine_reviews = pd.read_csv("wine_reviews.csv")
```

# Scatter plot

```python
sns.scatterplot(data=wine_reviews, x="points", y="price")
plt.title("Scatter Plot of Points vs Price")
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=wine_reviews, x="points", bins=20, kde=True)  # kde is optional here
plt.title("Histogram of Points")
plt.show()
```

# Bar chart

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=wine_reviews, x="country", order=wine_reviews["country"].value_counts().index[:10])
plt.title("Top 10 Countries with Most Wine Reviews")
plt.xticks(rotation=45)
plt.show()
```

# Heatmap

```python
plt.figure(figsize=(10, 8))
heatmap_data = wine_reviews.corr()
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Box plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=wine_reviews, x="points", y="country")
plt.title("Box Plot of Points by Country")
plt.xticks(rotation=45)
plt.show()
```
